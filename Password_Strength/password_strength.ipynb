```python
password_arr=np.array(data)
```

```python
password_arr
```

```python
import random
random.shuffle(password_arr)
```

```python
x=[labels[0] for labels in password_arr]
y=[labels[1] for labels in password_arr]
```

```python
x
```

```python
def word_break(inputs):
    charcter=[]
    for i in inputs:
        charcter.append(i)
    return charcter
```

```python
word_break('kzde5577')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer=TfidfVectorizer(tokenizer=word_break)
```

```python
data_X=vectorizer.fit_transform(x)
```

```python
data_X.shape
```

```python
vectorizer.get_feature_names()
```

```python
first_doc_vector=data_X[0]
first_doc_vector
```

```python
first_doc_vector.T.todense()
```

```python
df=pd.DataFrame(first_doc_vector.T.todense(),index=vectorizer.get_feature_names(),columns=['TF-IDF'])
df.sort_values(by=['TF-IDF'],ascending=False)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(data_X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf=LogisticRegression(random_state=0,multi_class='multinomial')
```

```python
clf.fit(X_train,Y_train)
```

```python
#Here we are giving our own password to check the stength of the password
dt=np.array(['Smartsai@123'])
pred=vectorizer.transform(dt)
clf.predict(pred)
```

```python
y_pred=clf.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm=confusion_matrix(Y_test,y_pred)
print(cm)
print(accuracy_score(Y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,y_pred))
```
